# Sample Analysis - a Comparison of Hot-Stints
The following is just one way you might want to use the scripts. In this example I completed three hot-stints in identical conditions at the Hungaroring to compare three different cars (the Lamborghini Hurac√†n EVO2, Ferrari 296 GT3 and Mercedes AMG GT3 Evo 2020) using the default aggressive setups.

## Setup

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Chart sizes
chart_size_x = 12
chart_size_y = 8
```

```python
# Import data from the Lamborghini stint
lamborghini = pd.read_csv("ACC-Data-2023-04-24-19-14-54.csv", index_col=0)
lamborghini = lamborghini[lamborghini["BestLapTime"] != 2147483647]
# Import data from the Ferrari stint
ferrari = pd.read_csv("ACC-Data-2023-04-24-19-31-10.csv", index_col=0)
ferrari = ferrari[ferrari["BestLapTime"] != 2147483647]
# Import data from the Mercedes stint
mercedes = pd.read_csv("ACC-Data-2023-04-24-19-57-07.csv", index_col=0)
mercedes = mercedes[mercedes["BestLapTime"] != 2147483647]
```

## Lamborghini Stats

```python
# Lamborghini tyre pressure deviation
x_axis = lamborghini["LapsCompleted"] + lamborghini["TrackPosition"]
target_psi = 26.5
y_axis_1, y_axis_2, y_axis_3, y_axis_4  = lamborghini["TyrePressureFrontLeft"] - target_psi, lamborghini["TyrePressureFrontRight"] - target_psi, lamborghini["TyrePressureRearLeft"] - target_psi, lamborghini["TyrePressureRearRight"] - target_psi
plt.figure(figsize=(chart_size_x,chart_size_y))
plt.ylim(-2, 2)
plt.axhline(y=0, color="black", linestyle="solid", label="Pressure Target")
plt.axhline(y=1, color="red", linestyle="--", label="Max. Pressure")
plt.axhline(y=-1, color="blue", linestyle="--", label="Min. Pressure")
plt.scatter(x_axis, y_axis_1, s=1, label="Front Left")
plt.scatter(x_axis, y_axis_2, s=1, label="Front Right")
plt.scatter(x_axis, y_axis_3, s=1, label="Rear Left")
plt.scatter(x_axis, y_axis_4, s=1, label="Rear Right")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
# Lamborghini tyre temperatures
x_axis = lamborghini["LapsCompleted"] + lamborghini["TrackPosition"]
y_axis_1, y_axis_2, y_axis_3, y_axis_4  = lamborghini["TyreTemperatureFrontLeft"], lamborghini["TyreTemperatureFrontRight"], lamborghini["TyreTemperatureRearLeft"], lamborghini["TyreTemperatureRearRight"]
plt.figure(figsize=(chart_size_x,chart_size_y))
plt.ylim(55, 115)
plt.axhline(y=105, color="red", linestyle="--", label="Max. Temperature")
plt.axhline(y=65, color="blue", linestyle="--", label="Min. Temperature")
plt.scatter(x_axis, y_axis_1, s=1, label="Front Left")
plt.scatter(x_axis, y_axis_2, s=1, label="Front Right")
plt.scatter(x_axis, y_axis_3, s=1, label="Rear Left")
plt.scatter(x_axis, y_axis_4, s=1, label="Rear Right")
plt.legend(loc="upper left")
plt.grid()
plt.show()
```

```python
lamborghini.agg({"TyrePressureFrontLeft": ["describe"], "TyrePressureFrontRight": ["describe"], "TyrePressureRearLeft": ["describe"], "TyrePressureRearRight": ["describe"], "TyreTemperatureFrontLeft": ["describe"], "TyreTemperatureFrontRight": ["describe"], "TyreTemperatureRearLeft": ["describe"], "TyreTemperatureRearRight": ["describe"], "TrackTemperature": ["describe"], "AmbientTemperature": ["describe"]})
```
